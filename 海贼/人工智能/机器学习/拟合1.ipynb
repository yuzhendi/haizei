```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
%pylab inline
print(tf.__version__)
```

```python
def plot_points(x, y, title_name):
    plt.title(title_name)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.scatter(x, y)
    plt.show()
```

```python
train_X = np.array([30.0, 40.0, 60.0,40.0, 80.0, 100.0,390.0])
train_Y = np.array([300.0, 310.0, 340.0, 330.0, 360.0,470.0, 400.0])
train_X /= 100.0
train_Y /= 100.0
plot_points(train_X, train_Y, title_name='Training Points')
```

```python
x = tf.placeholder(tf.float32, None)
y = tf.placeholder(tf.float32, None)
w = tf.Variable(np.random.randn(), name='weight') #变量
b = tf.Variable(np.random.randn(), name='bias') #变量
y_pred = tf.add(tf.multiply(w, x), b)
print(w)
```

```python
n_samples = train_X.shape[0]
cost = tf.reduce_sum(tf.pow((y_pred - y), 2)) / n_samples # 成本函数
learning_rate = 0.01 #学习速率
optimizer = tf.train.GradientDescentOptimizer(learning_rate).minimize(cost) #优化器
```

```python
training_epoch = 1000
display_step = 50
```

```python
with tf.Session() as sess:
    sess.run(tf.initialize_all_variables())
    for epoch in range(training_epoch):
        for (x_train, y_train) in zip(train_X, train_Y):
            loss = sess.run([optimizer,cost], feed_dict={x : x_train, y : y_train})
            print(loss)
```
